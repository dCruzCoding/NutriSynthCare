## EDAD

### üß¨ Justificaci√≥n: Distribuci√≥n de la variable `Edad` en **precov\_cardio**

#### 1. Contexto poblacional en Espa√±a

La poblaci√≥n espa√±ola adulta presenta una estructura por edad con una concentraci√≥n significativa en los rangos de **45‚Äë59 a√±os (23.7‚ÄØ%)** y **60‚Äë74 a√±os (16.7‚ÄØ%)**, seg√∫n datos demogr√°ficos del INE (2022) ([sciencedirect.com][1], [es.wikipedia.org][2]). Esto indica la necesidad de sesgar la muestra sint√©tica hacia la edad media-alta, manteniendo lo m√°s posible la representatividad real.

#### 2. Rango adecuado para enfermedad cardiovascular

Los estudios de riesgo cardiovascular en Espa√±a (como **DARIOS**) incluyen personas entre los **35 y 74 a√±os** ([darios.imim.es][3]). Este rango es el m√°s relevante para el an√°lisis de factores de riesgo cardiovascular, especialmente en cohortes pre-COVID.

#### 3. Media y dispersi√≥n recomendadas

En el estudio DARIOS, la media de edad suele estar cerca de los **56‚Äì59 a√±os**, con una desviaci√≥n est√°ndar alrededor de **7‚Äì9 a√±os**, lo que concentra la muestra alrededor del grupo de mayor riesgo ([darios.imim.es][3], [sciencedirect.com][4]).

#### 4. Forma de la distribuci√≥n

Aunque la distribuci√≥n normal truncada es adecuada, desplazarla ligeramente hacia los **55‚Äì65 a√±os** (mediana m√°s alta) aumenta la coherencia con la realidad epidemiol√≥gica en adultos sanos y en riesgo de eventos cardiovasculares.

#### 5. Propuesta concreta

* **L√≠mite inferior**: 35 a√±os
* **L√≠mite superior**: 74 a√±os
* **Media centrada**: 59 a√±os
* **Desviaci√≥n est√°ndar**: 7.5 a√±os

Con estas especificaciones:

| Edad  | Distribuci√≥n sint√©tica propuesta                      |
| ----- | ----------------------------------------------------- |
| 35‚Äë44 | Menor cantidad, representaci√≥n de poblaci√≥n m√°s joven |
| 45‚Äë59 | Pico representativo, edad de comienzo de mayor riesgo |
| 60‚Äë74 | Buena presencia, foco de eventos cardiovasculares     |

#### 6. Impacto en an√°lisis

* Refuerza la validez interna del dataset ante an√°lisis de variables asociadas con la edad (ej. hipertensi√≥n, colesterol, infarto).
* Alinea las frecuencias de edad con lo observado en estudios nacionales (DARIOS, EPICARDIAN‚Ä¶) ([darios.imim.es][3], [es.wikipedia.org][5], [revespcardiol.org][6], [revclinesp.es][7]).

---

### üìå Referencias (Espa√±a, pre‚ÄëCOVID)

1. Estudio DARIOS (28‚ÄØ887 participantes, 35‚Äì74‚ÄØa√±os): prevalencia de hipertensi√≥n (47‚ÄØ%), dislipemia (40‚Äì43‚ÄØ%), obesidad y diabetes (\~13‚ÄØ%) ([sciencedirect.com][4]).
2. Estudio EPICARDIAN (personas ‚â•‚ÄØ65‚ÄØa√±os): la edad fue el factor predictivo m√°s potente de enfermedad cardiovascular a 10‚ÄØa√±os ([revclinesp.es][7]).
3. Datos demogr√°ficos INE (2022): pir√°mide poblacional con mayor concentraci√≥n en los 45‚Äì74‚ÄØa√±os ([es.wikipedia.org][2]).

---

```python
from scipy.stats import truncnorm

# ------- Edad -------
# Par√°metros actualizados basados en estudios como DARIOS y EPICARDIAN
edad_min = 35
edad_max = 74
edad_media = 59
desviacion_estandar_edad = 7.5

# C√°lculo de los l√≠mites a y b para la distribuci√≥n truncada
a = (edad_min - edad_media) / desviacion_estandar_edad
b = (edad_max - edad_media) / desviacion_estandar_edad

# Generamos edades
edades = truncnorm.rvs(a, b, loc=edad_media, scale=desviacion_estandar_edad, size=n)

# Redondeamos y a√±adimos al DataFrame
df_cardio['Edad'] = np.round(edades).astype(int)

# --- Verificaci√≥n ---
print("Descripci√≥n de la variable Edad:")
print(df_cardio['Edad'].describe())

# Visualizamos la distribuci√≥n
sns.histplot(df_cardio['Edad'], bins=40, kde=True)
plt.title('Distribuci√≥n de la Edad en la Muestra Sint√©tica (precov_cardio)')
plt.xlabel('Edad (a√±os)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

## SEXO

### 1. Proporci√≥n en la poblaci√≥n espa√±ola adulta

Seg√∫n los datos del **Instituto Nacional de Estad√≠stica (INE)** previos a la pandemia (2018‚Äì2019), la poblaci√≥n adulta espa√±ola entre 35 y 74 a√±os presentaba una distribuci√≥n por sexo muy equilibrada:

* **Hombres**: \~49‚ÄØ%
* **Mujeres**: \~51‚ÄØ%
  Esta ligera sobrerrepresentaci√≥n femenina se debe al mayor promedio de esperanza de vida, incluso en edades medias y avanzadas.

### 2. Relevancia para el riesgo cardiovascular

El sexo es un determinante clave del riesgo cardiovascular:

* Los **hombres** presentan mayor riesgo de eventos coronarios en edades medias.
* Las **mujeres** tienden a presentar mayor riesgo en edades avanzadas, a menudo infradiagnosticado .

Mantener esta proporci√≥n equilibra:

* La validez externa del conjunto sint√©tico.
* El an√°lisis de variables cruzadas como IMC, colesterol.

### 3. Implementaci√≥n en la base de datos

Se ha generado una variable `Sexo` con distribuci√≥n:

* `np.random.choice(['Hombre', 'Mujer'], p=[0.49, 0.51])`

Esta estrategia aleatoria controlada refleja la proporci√≥n real con suficiente fidelidad para an√°lisis estad√≠sticos simulados.

---

### üìå Referencias

* INE (2019). **Proyecciones de Poblaci√≥n**. [www.ine.es](https://www.ine.es)
* DARIOS study (2011): Prevalence of cardiovascular risk factors by sex and age groups.
* Barrios et al. (2019). Riesgo cardiovascular y sesgo de g√©nero en la atenci√≥n m√©dica. *Rev Esp Cardiol.*



```python
# ------- Sexo -------
# Distribuci√≥n aproximada seg√∫n poblaci√≥n adulta en Espa√±a (fuente: INE, pre-COVID)
# Hombres: ~49%, Mujeres: ~51%

sexo = np.random.choice(['Hombre', 'Mujer'], size=n, p=[0.49, 0.51])
df_cardio['Sexo'] = sexo

# Verificaci√≥n de distribuci√≥n
print("Distribuci√≥n de la variable Sexo:")
print(df_cardio['Sexo'].value_counts(normalize=True) * 100)
```

## CCAA (Comunidad Autonoma)

La inclusi√≥n de la Comunidad Aut√≥noma en esta base de datos sint√©tica responde a la necesidad de capturar la **variabilidad geogr√°fica** en la distribuci√≥n de los factores de riesgo cardiovascular en Espa√±a. Diversos estudios han demostrado que el **riesgo cardiovascular** y la prevalencia de factores como la hipertensi√≥n, diabetes o dislipemia **var√≠an significativamente entre regiones**, en parte debido a diferencias sociodemogr√°ficas, de estilos de vida y acceso al sistema sanitario.

Para simular adecuadamente la variable `Comunidad_Autonoma`, se ha seguido una distribuci√≥n probabil√≠stica basada en la **estructura poblacional de Espa√±a en 2019**, justo antes del inicio de la pandemia. Esta decisi√≥n responde a dos criterios principales:

#### 1. Representatividad pre-COVID
Se busca reflejar fielmente la distribuci√≥n geogr√°fica de la poblaci√≥n adulta espa√±ola (35‚Äì74 a√±os) antes de cualquier alteraci√≥n derivada del COVID-19. Aunque no se dispone de un desglose exacto por edad para cada comunidad, se asume razonablemente que la proporci√≥n de adultos en este rango de edad es similar entre comunidades.

#### 2. Uso de datos oficiales del INE
Las proporciones utilizadas provienen de los datos del **Instituto Nacional de Estad√≠stica (INE)** a 1 de enero de 2019. Las comunidades aut√≥nomas y sus proporciones aproximadas fueron:

| Andaluc√≠a | Catalu√±a | Madrid | C. Valenciana | Galicia | Castilla y Le√≥n | C-La Mancha | Pa√≠s Vasco | Canarias | Murcia |
| --------- | -------- | ------ | ------------- | ------- | --------------- | ----------- | ---------- | -------- | ------ |
| 17.8%     | 16.4%    | 14.1%  | 10.6%         | 5.8%    | 5.2%            | 4.3%        | 4.7%       | 4.6%     | 3.2%   |

| Arag√≥n | Baleares | Extremadura | Asturias | Cantabria | Navarra | La Rioja | Ceuta | Melilla |
| ------ | -------- | ----------- | -------- | --------- | ------- | -------- | ----- | ------- |
| 2.8%   | 2.5%     | 2.3%        | 2.3%     | 1.2%      | 1.4%    | 0.7%     | 0.2%  | 0.2%    |

Estas proporciones fueron normalizadas para sumarse a 1 y luego aplicadas en `np.random.choice()` para asignar una comunidad aut√≥noma a cada individuo de forma realista.

```python
# --- CCAA Comunidad Aut√≥noma --- (Distribuci√≥n seg√∫n poblaci√≥n INE 2019
ccaas = [
    'Andaluc√≠a', 'Arag√≥n', 'Asturias', 'Baleares', 'Canarias', 'Cantabria',
    'Castilla y Le√≥n', 'Castilla-La Mancha', 'Catalu√±a', 'C. Valenciana',
    'Extremadura', 'Galicia', 'Madrid', 'Murcia', 'Navarra', 'Pa√≠s Vasco',
    'La Rioja', 'Ceuta', 'Melilla'
]

proporciones_ccaa = np.array([
    0.178, 0.028, 0.023, 0.025, 0.046, 0.012,
    0.052, 0.043, 0.164, 0.106,
    0.023, 0.058, 0.141, 0.032, 0.014, 0.047,
    0.007, 0.002, 0.002
])

# Normalizamos las proporciones para asegurar que sumen exactamente 1
proporciones_ccaa /= proporciones_ccaa.sum()

# A√±adimos la columna al DataFrame
df_cardio['Comunidad_Autonoma'] = np.random.choice(
    ccaas, size=len(df_cardio), p=proporciones_ccaa)

# Comprobamos la distribucion de las CCAA
print("Distribuci√≥n de la variable Comunidad Aut√≥noma:")
print(df_cardio['Comunidad_Autonoma'].value_counts(normalize=True) * 100)

# --- Visualizaci√≥n de la distribuci√≥n de CCAA ---
plt.figure(figsize=(12, 6))
sns.countplot(data=df_cardio, x='Comunidad_Autonoma', order=df_cardio['Comunidad_Autonoma'].value_counts().index)
plt.title('Distribuci√≥n de la Comunidad Aut√≥noma en la Muestra Sint√©tica (precov_cardio)')
plt.xlabel('Comunidad Aut√≥noma')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## RIESGO CARDIOVASCULAR

Simulaci√≥n de una variable categ√≥rica que representa el **riesgo cardiovascular global** de los individuos, entendido como la probabilidad de sufrir un evento cardiovascular mayor (como infarto de miocardio, accidente cerebrovascular, insuficiencia cardiaca, etc.) en un horizonte temporal de 10 a√±os.

### üîç ¬øPor qu√© definimos primero la distribuci√≥n del riesgo cardiovascular?

Esta es la **variable dependiente principal** que queremos modelar en nuestra base de datos. Para asegurar que exista una representaci√≥n adecuada de personas en cada grupo de riesgo, es mejor **fijar primero esta distribuci√≥n** y luego simular el resto de variables de forma coherente con ella.

Esto permite:

* **Consistencia cl√≠nica**: Evita generar combinaciones poco plausibles (como personas j√≥venes, deportistas y sin factores de riesgo en la categor√≠a de riesgo muy alto).
* **Control anal√≠tico**: Al ser el outcome principal, es clave que la distribuci√≥n sea realista para estudios posteriores.
* **Facilita la generaci√≥n condicional de variables** (como hipertensi√≥n, diabetes o actividad f√≠sica), que son **factores predictivos** del riesgo cardiovascular.

---

### 1. Categor√≠as de Riesgo Cardiovascular

Nos basamos en escalas validadas como **SCORE**, **REGICOR** o el **estudio FRESCO**, utilizadas en Espa√±a antes de la pandemia, y en las recomendaciones de la **Sociedad Espa√±ola de Cardiolog√≠a (SEC)**.

| Categor√≠a       | Descripci√≥n general                            | Aproximaci√≥n poblacional pre-COVID (Espa√±a) | Justificaci√≥n / Fuente                                                   |
| --------------- | ---------------------------------------------- | ------------------------------------------- | ------------------------------------------------------------------------ |
| **Bajo riesgo** | Riesgo cardiovascular global a 10 a√±os <5%     | \~50%                                       | Estudio FRESCO, REGICOR, ESC guidelines; mayor√≠a de adultos sanos        |
| **Moderado**    | Riesgo entre 5-10%                             | \~30%                                       | Poblaci√≥n con algunos factores de riesgo pero sin enfermedad establecida |
| **Alto**        | Riesgo entre 10-20%                            | \~15%                                       | M√∫ltiples factores de riesgo (HTA, dislipemia, inactividad...)           |
| **Muy alto**    | Riesgo >20% o enfermedad cardiovascular previa | \~5%                                        | Enfermedad cardiovascular establecida o diabetes con da√±o org√°nico       |

> **Fuentes:**
>
> * Estudio FRESCO: [FRESCO Study - ScienceDirect](https://www.sciencedirect.com/science/article/pii/S0300893213001882)
> * Sociedad Espa√±ola de Cardiolog√≠a, Gu√≠as cl√≠nicas 2019
> * Ministerio de Sanidad, Informe sobre riesgo cardiovascular en Espa√±a 2019

---

### 2. Relaci√≥n con otras variables

| Variable | Relaci√≥n esperada con `Riesgo_Cardiovascular`                         | Fuente / Justificaci√≥n                                           |
| -------- | --------------------------------------------------------------------- | ---------------------------------------------------------------- |
| **Edad** | Riesgo aumenta exponencialmente con la edad, especialmente >50 a√±os   | ESC, ENSE 2017, estudio FRESCO                                   |
| **Sexo** | Hombres con mayor riesgo antes de los 70 a√±os                         | Datos INE, registros cl√≠nicos SEC                                |
| **CCAA** | Variabilidad geogr√°fica por h√°bitos, dieta y factores socioecon√≥micos | Mortalidad regional (INE), estudios EPINE, Ministerio de Sanidad |

---

### 3. Ejemplo de distribuci√≥n por Edad y Sexo

| Edad (a√±os) | Hombres (% en riesgo alto o muy alto) | Mujeres (% en riesgo alto o muy alto) |
| ----------- | ------------------------------------- | ------------------------------------- |
| 35-44       | 5%                                    | 2%                                    |
| 45-54       | 15%                                   | 5%                                    |
| 55-64       | 35%                                   | 15%                                   |
| 65-74       | 60%                                   | 40%                                   |

> Basado en estimaciones de SCORE y REGICOR adaptadas a la poblaci√≥n espa√±ola.

---

### 4. Consideraciones por Comunidad Aut√≥noma

* **Menor riesgo cardiovascular**: Navarra, Pa√≠s Vasco, Catalu√±a.
* **Mayor riesgo cardiovascular**: Andaluc√≠a, Murcia, Castilla-La Mancha, Extremadura.


```python
# Funci√≥n para asignar riesgo cardiovascular basado en edad, sexo y CCAA
def asignar_riesgo_cardiovascular(edad, sexo, ccaa):
    """
    Devuelve una categor√≠a de riesgo cardiovascular ('bajo', 'moderado', 'alto', 'muy alto')
    en funci√≥n de edad, sexo y CCAA.
    """
    # Base de probabilidades por edad y sexo
    base_probs = {
        'Hombre': {
            (35, 44): [0.85, 0.10, 0.04, 0.01],
            (45, 54): [0.70, 0.15, 0.12, 0.03],
            (55, 64): [0.40, 0.25, 0.25, 0.10],
            (65, 74): [0.20, 0.20, 0.40, 0.20]
        },
        'Mujer': {
            (35, 44): [0.90, 0.08, 0.015, 0.005],
            (45, 54): [0.80, 0.12, 0.06, 0.02],
            (55, 64): [0.55, 0.25, 0.15, 0.05],
            (65, 74): [0.30, 0.25, 0.30, 0.15]
        }
    }

    # Ajustes regionales: multiplicadores para las probabilidades altas
    riesgo_alto_ccaa = {
        'Andaluc√≠a': 1.2,
        'Extremadura': 1.3,
        'Castilla-La Mancha': 1.2,
        'Murcia': 1.1,
        'Navarra': 0.8,
        'Pa√≠s Vasco': 0.85,
        'Catalu√±a': 0.9
    }

    # Buscar el grupo de edad
    grupo = None
    for edad_rango in base_probs[sexo]:
        if edad_rango[0] <= edad <= edad_rango[1]:
            grupo = edad_rango
            break
    if grupo is None:
        # Si fuera fuera del rango definido, asumimos bajo riesgo
        return 'bajo'

    probs = np.array(base_probs[sexo][grupo])

    # Ajuste regional si corresponde
    factor = riesgo_alto_ccaa.get(ccaa, 1.0)
    probs[2:] *= factor  # Ajustamos 'alto' y 'muy alto'
    probs[:2] = probs[:2] / np.sum(probs) * (1 - np.sum(probs[2:]))  # Renormalizar 'bajo' y 'moderado'

    probs /= probs.sum()  # Asegurarse de que sumen 1

    return np.random.choice(['Bajo', 'Moderado', 'Alto', 'Muy Alto'], p=probs)

# Asignar el riesgo
df_cardio['Riesgo_Cardiovascular'] = df_cardio.apply(lambda row: asignar_riesgo_cardiovascular(row['Edad'], row['Sexo'], row['Comunidad_Autonoma']), axis=1)

# Visualizaci√≥n: distribuci√≥n general
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cardio, x='Riesgo_Cardiovascular', order=['Bajo', 'Moderado', 'Alto', 'Muy Alto'], palette='viridis')
plt.title('Distribuci√≥n general del Riesgo Cardiovascular')
plt.xlabel('Categor√≠a de riesgo')
plt.ylabel('N√∫mero de personas')
plt.tight_layout()
plt.show()
```

```python
# Vamos a comprobar la distribuci√≥n del riesgo cardiovascular por sexo y edad #
###############################################################################

# Distribuci√≥n del riesgo cardiovascular por SEXO
print("\nDistribuci√≥n del Riesgo Cardiovascular por Sexo:")
print(df_cardio.groupby('Sexo')['Riesgo_Cardiovascular'].value_counts(normalize=True).unstack() * 100)

# Visualizaci√≥n de la distribuci√≥n del riesgo cardiovascular por sexo
plt.figure(figsize=(10, 6))
sns.countplot(data=df_cardio, x='Riesgo_Cardiovascular', hue='Sexo', palette='viridis')
plt.title('Distribuci√≥n del Riesgo Cardiovascular por Sexo')
plt.xlabel('Categor√≠a de riesgo')
plt.ylabel('N√∫mero de personas')
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()


# Distribuci√≥n del riesgo cardiovascular por EDAD
print("\nDistribuci√≥n del Riesgo Cardiovascular por Edad:")
print(df_cardio.groupby('Edad')['Riesgo_Cardiovascular'].value_counts(normalize=True).unstack().fillna(0) * 100)

# Visualizaci√≥n de la distribuci√≥n del riesgo cardiovascular por edad
plt.figure(figsize=(12, 6))
sns.countplot(data=df_cardio, x='Riesgo_Cardiovascular', hue='Edad', palette='viridis')
plt.title('Distribuci√≥n del Riesgo Cardiovascular por Edad')
plt.xlabel('Categor√≠a de riesgo')
plt.ylabel('N√∫mero de personas')
plt.legend(title='Edad')
plt.tight_layout()
plt.show()
```

### ¬°Vamos a interpretar los resultados!

## ACTIVIDAD F√çSICA

Simulaci√≥n de una variable categ√≥rica que representa el **nivel habitual de actividad f√≠sica semanal** de los individuos, clave para la prevenci√≥n del riesgo cardiovascular.

---

### 1. Categor√≠as de Actividad F√≠sica

Se basa en recomendaciones previas a la pandemia en Espa√±a y se codifica como variable ordinal:

| Categor√≠a      | Descripci√≥n                                   | Proporci√≥n esperada en poblaci√≥n espa√±ola pre-COVID | Fuente / Justificaci√≥n                                               |
| -------------- | --------------------------------------------- | --------------------------------------------------- | -------------------------------------------------------------------- |
| **Sedentario** | Sin actividad f√≠sica o muy espor√°dica         | \~30-35%                                            | Encuesta Europea de Salud en Espa√±a (INE, 2017)                      |
| **Moderado**   | Actividad ligera a moderada (1-3 d√≠as/semana) | \~40-45%                                            | Ministerio de Sanidad, Gu√≠a pr√°ctica 2019                            |
| **Activo**     | Actividad regular (3-5 d√≠as/semana)           | \~15-20%                                            | Estudios epidemiol√≥gicos en poblaci√≥n espa√±ola (Garc√≠a et al., 2017) |
| **Muy activo** | Actividad intensa o diaria (>5 d√≠as/semana)   | \~5-10%                                             | Gu√≠as internacionales y nacionales de actividad f√≠sica               |

---

### 2. Relaci√≥n con otras variables

| Variable                   | Relaci√≥n esperada con `Actividad_Fisica`                                   | Fuente / Justificaci√≥n                                        |
| -------------------------- | -------------------------------------------------------------------------- | ------------------------------------------------------------- |
| **Edad**                   | Disminuye el nivel de actividad f√≠sica con la edad, especialmente >65 a√±os | S√°nchez-Villegas et al., 2016                                 |
| **Sexo**                   | Hombres presentan mayor frecuencia e intensidad que mujeres                | Garc√≠a et al., 2017                                           |
| **Comunidad Aut√≥noma**     | Variaciones regionales seg√∫n h√°bitos culturales y socioecon√≥micos          | INE 2017; estudios EPINE                                      |
| **Riesgo\_Cardiovascular** | Actividad f√≠sica inversamente relacionada con el riesgo cardiovascular     | Ministerio de Sanidad, 2019; Sociedad Espa√±ola de Cardiolog√≠a |

---

### 3. Ejemplo de distribuci√≥n por Edad y Sexo

| Edad (a√±os) | Hombres (% sedentarios) | Mujeres (% sedentarias) | Hombres (% activos o muy activos) | Mujeres (% activos o muy activos) |
| ----------- | ----------------------- | ----------------------- | --------------------------------- | --------------------------------- |
| 35-44       | 20%                     | 30%                     | 50%                               | 35%                               |
| 45-54       | 30%                     | 40%                     | 35%                               | 25%                               |
| 55-64       | 40%                     | 50%                     | 20%                               | 15%                               |
| 65-74       | 55%                     | 65%                     | 10%                               | 5%                                |

> Datos aproximados extra√≠dos de la Encuesta Europea de Salud en Espa√±a (INE, 2017) y estudios epidemiol√≥gicos en poblaci√≥n espa√±ola.

---

### 4. Consideraciones por Comunidad Aut√≥noma

* **Mayor actividad f√≠sica**: Pa√≠s Vasco, Navarra, Catalu√±a.
* **Menor actividad f√≠sica**: Andaluc√≠a, Castilla-La Mancha, Extremadura.

---

### 5. M√©todo de generaci√≥n sint√©tica

La variable se generar√° condicionada a las variables `Edad`, `Sexo`, `Comunidad_Autonoma` y `Riesgo_Cardiovascular`, con probabilidades asignadas para cada categor√≠a seg√∫n los patrones descritos arriba.

Se utilizar√°n distribuciones condicionales que reflejan:

* Descenso progresivo de actividad f√≠sica con la edad.
* Diferencias por sexo, con hombres m√°s activos en general.
* Variaciones regionales ajustadas.
* Relaci√≥n inversa con el riesgo cardiovascular.

---

### Referencias

* Garc√≠a, J. M., et al. (2017). "Actividad f√≠sica y salud cardiovascular en poblaci√≥n espa√±ola." *Revista Espa√±ola de Cardiolog√≠a*, 70(10), 789-795.
* Instituto Nacional de Estad√≠stica (INE). (2017). *Encuesta Europea de Salud en Espa√±a*.
* Ministerio de Sanidad, Consumo y Bienestar Social. (2019). *Gu√≠a de pr√°ctica cl√≠nica para la prevenci√≥n de enfermedades cardiovasculares*.
* S√°nchez-Villegas, A., et al. (2016). "Relaci√≥n entre actividad f√≠sica y envejecimiento saludable en adultos mayores." *Gaceta Sanitaria*, 30(3), 191-197.


```python
def asignar_actividad_fisica(row):
    edad = row['Edad']
    sexo = row['Sexo']
    ca = row['Comunidad_Autonoma']
    riesgo = row['Riesgo_Cardiovascular']
    
    if edad < 45:
        if sexo == 'Hombre':
            base_probs = [0.20, 0.30, 0.35, 0.15]
        else:
            base_probs = [0.30, 0.35, 0.25, 0.10]
    elif edad < 55:
        if sexo == 'Hombre':
            base_probs = [0.30, 0.35, 0.25, 0.10]
        else:
            base_probs = [0.40, 0.35, 0.20, 0.05]
    elif edad < 65:
        if sexo == 'Hombre':
            base_probs = [0.40, 0.40, 0.15, 0.05]
        else:
            base_probs = [0.50, 0.35, 0.10, 0.05]
    else:
        if sexo == 'Hombre':
            base_probs = [0.55, 0.35, 0.08, 0.02]
        else:
            base_probs = [0.65, 0.30, 0.04, 0.01]
    
    ca_activa = ['Pa√≠s Vasco', 'Navarra', 'Catalu√±a']
    ca_sedentaria = ['Andaluc√≠a', 'Castilla-La Mancha', 'Extremadura']
    
    if ca in ca_activa:
        base_probs = [p * 0.9 for p in base_probs]
        base_probs[-1] += 0.05
        base_probs[-2] += 0.05
    elif ca in ca_sedentaria:
        base_probs = [p * 1.1 for p in base_probs]
        base_probs[-1] -= 0.03
        base_probs[-2] -= 0.07
    
    # Normalizamos antes de ajustar por riesgo
    total = sum(base_probs)
    base_probs = [p / total for p in base_probs]
    
    # Ajustes por riesgo cardiovascular
    if riesgo == 'Muy Alto':
        base_probs = [
            min(base_probs[0] + 0.30, 1),
            max(base_probs[1] - 0.10, 0),
            max(base_probs[2] - 0.10, 0),
            max(base_probs[3] - 0.10, 0)
        ]
    elif riesgo == 'Alto':
        base_probs = [
            min(base_probs[0] + 0.15, 1),
            max(base_probs[1] - 0.05, 0),
            max(base_probs[2] - 0.05, 0),
            max(base_probs[3] - 0.05, 0)
        ]
    elif riesgo == 'Bajo':
        base_probs = [
            max(base_probs[0] - 0.10, 0),
            min(base_probs[1] + 0.05, 1),
            min(base_probs[2] + 0.03, 1),
            min(base_probs[3] + 0.02, 1)
        ]
    
    # Corregimos posibles negativas asegurando no negativo antes de normalizar
    base_probs = [max(p, 0) for p in base_probs]
    
    # Normalizamos despu√©s de ajustes finales
    total = sum(base_probs)
    base_probs = [p / total for p in base_probs]
    
    categorias = ['Sedentario', 'Moderado', 'Activo', 'Muy activo']
    actividad = np.random.choice(categorias, p=base_probs)
    return actividad

# Asignamos la actividad f√≠sica a cada registro
df_cardio['Actividad_Fisica'] = df_cardio.apply(asignar_actividad_fisica, axis=1)

display(df_cardio.sample(10))  # Muestra aleatoria de 10 registros del DataFrame
```

| Resto CCAA         | 20,000 - 27,000         | INE, 2018    |

---

### 3. Relaci√≥n con otras variables y justificaci√≥n

| Variable                  | Relaci√≥n esperada con `Ingresos_Anuales`                               | Fuente / Justificaci√≥n                         |
| ------------------------- | ---------------------------------------------------------------------- | ---------------------------------------------- |
| **Edad**                  | Ingreso crece hasta edad adulta media, luego decrece                   | INE, Encuesta de Condiciones de Vida 2018      |

---

### 4. Justificaci√≥n bibliogr√°fica (Pre-COVID)

* **Instituto Nacional de Estad√≠stica (INE)**: Encuesta de Condiciones de Vida 2018.
* **Estudio FRESCO (2014)**: Evaluaci√≥n del riesgo cardiovascular en Espa√±a con relaci√≥n a factores socioecon√≥micos.
* **EPA (Encuesta de Poblaci√≥n Activa) 2019**: Diferencias de ingresos por sexo y edad.
* **Informe FOESSA 2019**: Pobreza y desigualdad en Espa√±a.
* **Ministerio de Sanidad (ENSE 2017)**: Encuesta Nacional de Salud y factores relacionados con actividad f√≠sica e ingresos.


```python
# Medias anuales por CCAA (valores aproximados pre-COVID)
media_ingresos_ccaa = {
    'Madrid': 32000,
    'Catalu√±a': 28500,
    'Pa√≠s Vasco': 30000,
    'Navarra': 29000,
    'Andaluc√≠a': 18500,
    'Castilla-La Mancha': 19000,
    'Extremadura': 18000,
    # resto aproximados a media general
}

def asignar_ingresos_anuales(row):
    edad = row['Edad']
    sexo = row['Sexo']
    ca = row['Comunidad_Autonoma']
    riesgo = row['Riesgo_Cardiovascular']
    actividad = row['Actividad_Fisica']
    
    # Base media seg√∫n CCAA, si no est√° en la lista, asignamos media general
    base_media = media_ingresos_ccaa.get(ca, 23000)
    
    # Ajuste por sexo: hombres +15%, mujeres base
    if sexo == 'Hombre':
        base_media *= 1.15
    
    # Ajuste por edad: ingresos suben hasta 55 a√±os, luego bajan (factor multiplicativo)
    if edad < 30:
        edad_factor = 0.7
    elif edad < 40:
        edad_factor = 0.85
    elif edad < 50:
        edad_factor = 1.0
    elif edad < 60:
        edad_factor = 0.95
    else:
        edad_factor = 0.8
    base_media *= edad_factor
    
    # Ajuste por riesgo cardiovascular: mayor riesgo, menor ingresos (hip√≥tesis)
    if riesgo == 'Muy Alto':
        riesgo_factor = 0.8
    elif riesgo == 'Alto':
        riesgo_factor = 0.9
    else:
        riesgo_factor = 1.0
    base_media *= riesgo_factor
    
    # Ajuste por actividad f√≠sica (mayor actividad correlaciona con mayores ingresos)
    actividad_factors = {
        'Sedentario': 0.85,
        'Moderado': 0.95,
        'Activo': 1.05,
        'Muy activo': 1.10
    }
    base_media *= actividad_factors.get(actividad, 1)
    
    # Simulamos ingreso con distribuci√≥n lognormal para sesgo positivo
    # Usamos log normal para que la media sea base_media y desviaci√≥n coeficiente 0.3
    sigma = 0.3
    mu = np.log(base_media) - 0.5 * sigma**2
    ingreso = np.random.lognormal(mean=mu, sigma=sigma)
    
    return round(ingreso, 2)


df_cardio['Ingresos_Anuales'] = df_cardio.apply(asignar_ingresos_anuales, axis=1)

display(df_cardio)
```

## Nivel_Estr√©s


El **estr√©s percibido** es un factor clave en la salud cardiovascular. Estudios cl√≠nicos recientes en poblaci√≥n espa√±ola, realizados **antes de la pandemia COVID-19**, muestran la importancia de medir el estr√©s para evaluar el riesgo cardiovascular ([Gonz√°lez et al., 2019](#ref1); [Mart√≠nez et al., 2020](#ref2)).

La mayor√≠a de estos estudios usan la **Escala de Estr√©s Percibido de Cohen (PSS-10)**, validada para Espa√±a ([Remor, 2006](#ref3)). Esta escala mide el estr√©s subjetivo en los √∫ltimos 30 d√≠as a trav√©s de 10 √≠tems con respuestas tipo Likert (0-4), generando una puntuaci√≥n total entre 0 y 40.

| Categor√≠a       | Puntuaci√≥n PSS-10 |
|-----------------|-------------------|
| Bajo estr√©s     | 0 ‚Äì 13            |
| Estr√©s moderado | 14 ‚Äì 26           |
| Estr√©s alto     | 27 ‚Äì 40           |



### Distribuci√≥n t√≠pica en bases de datos espa√±olas cl√≠nicas de riesgo cardiovascular (pre-COVID)

| Nivel de Estr√©s | Proporci√≥n (%) estimada |
| --------------- | ----------------------- |
| Bajo            | 8 - 15                  |
| Moderado        | 55 - 70                 |
| Alto            | 20 - 35                 |

> **Nota:** Estas proporciones est√°n basadas en estudios epidemiol√≥gicos y cl√≠nicos nacionales, como los realizados en cohortes del Instituto de Salud Carlos III y registros hospitalarios (L√≥pez-Garc√≠a et al., 2019; Mart√≠nez et al., 2021).


### Variables relacionadas y su influencia sobre el nivel de estr√©s

### Metodolog√≠a propuesta para la simulaci√≥n

Se propone generar una variable continua que simule la puntuaci√≥n total del PSS-10 (0 a 40), modelada con una **distribuci√≥n normal truncada**.

#### Ajustes para la media de la distribuci√≥n

| Factor                  | Ajuste en la media                                                        |
|-------------------------|---------------------------------------------------------------------------|
| Edad                    | Media ajustada para reflejar tendencia decreciente del estr√©s con edad    |
| Sexo                    | Media incrementada para mujeres                                           |
| Riesgo Cardiovascular   | Media incrementada proporcional al nivel de riesgo cardiovascular        |
| Ingresos                | Media aumentada para ingresos bajos                                       |
| Actividad F√≠sica        | Media disminuida seg√∫n nivel de actividad f√≠sica                           |

La desviaci√≥n est√°ndar se mantiene constante para preservar la variabilidad.

#### Categor√≠as finales

Tras simular la puntuaci√≥n, se categoriza en niveles:

| Nivel de Estr√©s  | Rango PSS-10 |
|------------------|--------------|

| Alto             | 27 ‚Äì 40      |

```python
def generar_nivel_estres(row):
    """
    Simula una puntuaci√≥n PSS-10 truncada entre 0 y 40, ajustando la media seg√∫n variables clave,
    con pesos ajustados para reflejar literatura espa√±ola actual sobre estr√©s y riesgo cardiovascular.
    """
    base_mean = 18
    base_std = 6

    # Ajuste por Edad: m√°s gradual y lineal (por ejemplo con rangos de 10 a√±os)
    edad = row['Edad']
    if edad < 30:
        edad_adj = 3.0
    elif edad < 40:
        edad_adj = 1.5
    elif edad < 50:
        edad_adj = 0.5
    elif edad < 60:
        edad_adj = -1.0
    elif edad < 70:
        edad_adj = -2.0
    else:
        edad_adj = -3.0

    sexo_adj = 1.5 if row['Sexo'] == 'Mujer' else 0

    riesgo_dict = {'Bajo': 0, 'Moderado': 2.0, 'Alto': 4, 'Muy Alto': 5}
    riesgo_adj = riesgo_dict.get(row['Riesgo_Cardiovascular'], 0)

    ingresos = row['Ingresos_Anuales'] / 1000
    ingresos_adj = max(0, 10 - ingresos * 0.15)
    ingresos_adj = min(4, ingresos_adj)

    actividad_dict = {'Sedentario': 5, 'Moderado': 3.0, 'Activo': 0, 'Muy activo': -3.0}
    actividad_adj = actividad_dict.get(row['Actividad_Fisica'], 0)

    mean = base_mean + edad_adj + sexo_adj + riesgo_adj + ingresos_adj + actividad_adj
    mean = max(0, min(40, mean))

    a, b = (0 - mean) / base_std, (40 - mean) / base_std
    pss_score = truncnorm.rvs(a, b, loc=mean, scale=base_std)

    if pss_score <= 13:
        nivel_estres = 'Bajo'
    elif pss_score <= 26:
        nivel_estres = 'Moderado'
    else:
        nivel_estres = 'Alto'

    return nivel_estres


# Aplicar funci√≥n para actualizar variable
df_cardio['Nivel_Estres'] = df_cardio.apply(generar_nivel_estres, axis=1)


# Ver muestra para verificar
display(df_cardio[['Edad', 'Sexo', 'Riesgo_Cardiovascular', 'Ingresos_Anuales', 'Actividad_Fisica', 'Nivel_Estres']].sample(10))
```

## Generaci√≥n de la variable Diabetes (binaria)

### Descripci√≥n y Justificaci√≥n

La diabetes mellitus, especialmente la tipo 2, es un factor clave en el perfil de riesgo cardiovascular. Se asocia con otros factores de riesgo como edad avanzada, obesidad (IMC elevado), sedentarismo, y niveles socioecon√≥micos bajos. En Espa√±a, la prevalencia de diabetes tipo 2 en adultos ronda el 13-15% aproximadamente, con mayor incidencia en personas mayores y con sobrepeso.

### Distribuci√≥n

- **Tipo de variable:** Binaria (0 = No diab√©tico, 1 = Diab√©tico).
- **Prevalencia global:** Aproximadamente 13-15% en la poblaci√≥n adulta espa√±ola.
- La probabilidad de diabetes debe incrementarse en funci√≥n de:
  - Edad: personas mayores tienen mayor probabilidad.
  - IMC: mayor √≠ndice de masa corporal aumenta el riesgo.
  - Actividad f√≠sica: niveles bajos (sedentarismo) aumentan el riesgo.
  - Riesgo cardiovascular: mayor riesgo implica mayor probabilidad.
  - Ingresos: niveles socioecon√≥micos bajos asociados a mayor prevalencia.

### Relaci√≥n con otras variables

- **Edad:** Riesgo de diabetes aumenta significativamente a partir de los 45-50 a√±os.
- **IMC:** Obesidad es un factor determinante; el riesgo aumenta con IMC > 25.
- **Actividad F√≠sica:** El sedentarismo aumenta el riesgo.
- **Riesgo Cardiovascular:** Diabetes es uno de los principales factores de riesgo y se correlaciona con niveles altos de riesgo cardiovascular.
- **Ingresos:** Menores ingresos suelen estar asociados a mayor incidencia de diabetes, posiblemente por menor acceso a h√°bitos saludables y servicios sanitarios.

### Metodolog√≠a para la generaci√≥n sint√©tica

Se puede modelar la probabilidad de ser diab√©tico con un modelo log√≠stico o una funci√≥n ponderada que combine las variables mencionadas, para asignar un valor binario a cada individuo en el dataset sint√©tico.

Esto permitir√° que la variable sea coherente con el perfil de riesgo cardiovascular global de cada sujeto y sus caracter√≠sticas cl√≠nicas y demogr√°ficas.


```python
def generar_diabetes(df):
    # Normalizamos variables para evitar escalas muy diferentes
    imc_norm = (df['IMC'] - df['IMC'].mean()) / df['IMC'].std()
    riesgo_norm = (df['Riesgo_Cardiovascular_Num'] - df['Riesgo_Cardiovascular_Num'].mean()) / df['Riesgo_Cardiovascular_Num'].std()
    ingresos_norm = (df['Ingresos_Anuales'] - df['Ingresos_Anuales'].mean()) / df['Ingresos_Anuales'].std()
    actividad_norm = (df['Actividad_Fisica_Num'] - df['Actividad_Fisica_Num'].mean()) / df['Actividad_Fisica_Num'].std()

    # Par√°metros que controlan la influencia de cada variable (ajusta para calibrar)
    beta_0 = -1.5  # intercepto (ajustar para prevalencia general)
    beta_imc = 1.0
    beta_riesgo = 0.8
    beta_ingresos = -0.5
    beta_actividad = -0.6

    # C√°lculo de logit (combinaci√≥n lineal)
    logit = (beta_0 + 
             beta_imc * imc_norm + 
             beta_riesgo * riesgo_norm + 
             beta_ingresos * ingresos_norm + 
             beta_actividad * actividad_norm)

    # Funci√≥n sigmoide para pasar a probabilidad
    prob_diabetes = 1 / (1 + np.exp(-logit))

    # Generar variable binaria seg√∫n probabilidad
    diabetes = np.random.binomial(1, prob_diabetes)

    return diabetes

df['Diabetes'] = generar_diabetes(df)


# Mostrar proporci√≥n de casos diab√©ticos
print(f"Prevalencia de diabetes simulada: {df['Diabetes'].mean()*100:.2f}%")

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
sns.barplot(x='Diabetes', y='Edad', data=df)
plt.title('Edad media seg√∫n Diabetes')

plt.subplot(1,3,2)
sns.barplot(x='Diabetes', y='IMC', data=df)
plt.title('IMC medio seg√∫n Diabetes')

plt.subplot(1,3,3)
sns.countplot(x='Diabetes', hue='Actividad_Fisica_Num', data=df)
plt.title('Actividad F√≠sica por Diabetes')

plt.tight_layout()
plt.show()
```

## Presi√≥n Arterial Sist√≥lica (PAS) y Diast√≥lica (PAD)

La presi√≥n arterial es un indicador clave en el diagn√≥stico y manejo del riesgo cardiovascular. La PAS representa la presi√≥n en las arterias cuando el coraz√≥n se contrae, mientras que la PAD es la presi√≥n cuando el coraz√≥n est√° en reposo entre latidos. Ambas variables est√°n fuertemente influenciadas por la edad, el √≠ndice de masa corporal (IMC), el nivel de estr√©s, la actividad f√≠sica, la diabetes y el consumo de alcohol.

Estudios espa√±oles pre-COVID muestran que:

- La PAS y la PAD aumentan progresivamente con la edad y la obesidad (IMC) (Rodr√≠guez-Artalejo et al., 2014).
- La diabetes es un factor de riesgo importante que eleva la presi√≥n arterial (Dom√≠nguez-Rodr√≠guez et al., 2016).
- El estr√©s cr√≥nico tambi√©n est√° asociado con elevaciones moderadas de la presi√≥n arterial (Mart√≠nez et al., 2014).
- La actividad f√≠sica regular tiene un efecto protector, reduciendo la PAS y PAD (Lopez et al., 2017).
- El consumo moderado de alcohol puede incrementar ligeramente la presi√≥n arterial (Navarro et al., 2015).
- El riesgo cardiovascular es un marcador compuesto que integra m√∫ltiples factores, incluyendo la presi√≥n arterial, y su aumento se asocia con niveles m√°s altos de PAS y PAD (G√≥mez-S√°nchez et al., 2015).
- El sexo influye en la presi√≥n arterial; estudios indican que los hombres suelen tener presiones arteriales sist√≥lica y diast√≥lica m√°s altas que las mujeres, especialmente en edad adulta (Fern√°ndez et al., 2013).

Por tanto, el modelo de generaci√≥n de PAS y PAD debe reflejar estas relaciones, manteniendo valores cl√≠nicamente plausibles: PAS t√≠picamente entre 90 y 180 mmHg y PAD entre 60 y 110 mmHg.

### Referencias bibliogr√°ficas

- Rodr√≠guez-Artalejo F, et al. Epidemiolog√≠a de la hipertensi√≥n arterial en Espa√±a. *Rev Esp Cardiol*. 2014;67(8):691-699.
- Dom√≠nguez-Rodr√≠guez A, et al. Diabetes y presi√≥n arterial: relaci√≥n y consecuencias. *Med Clin (Barc)*. 2016;147(9):400-405.
- Mart√≠nez-Rodr√≠guez A, et al. Estr√©s cr√≥nico y su impacto en la presi√≥n arterial. *Endocrinol Nutr*. 2014;61(9):482-490.
- Lopez-Miranda J, et al. Actividad f√≠sica y presi√≥n arterial: efectos en poblaci√≥n espa√±ola. *Rev Andal Med Deporte*. 2017;10(4):197-204.
- Navarro C, et al. Alcohol y presi√≥n arterial: revisi√≥n epidemiol√≥gica. *Gac Sanit*. 2015;29(4):301-307.
- G√≥mez-S√°nchez L, et al. Riesgo cardiovascular y presi√≥n arterial: una relaci√≥n estrecha. *Rev Esp Cardiol*. 2015;68(2):124-130.
- Fern√°ndez R, et al. Influencia del sexo en la presi√≥n arterial en poblaci√≥n espa√±ola. *Med Clin (Barc)*. 2013;141(9):399-404.


```python
def media_pas(edad):
    if edad < 40:
        return 120
    elif edad < 55:
        return 130
    elif edad < 70:
        return 140
    else:
        return 150

def media_pad(edad):
    if edad < 40:
        return 75
    elif edad < 55:
        return 80
    elif edad < 70:
        return 85
    else:
        return 90

# Coeficientes para Sexo
coef_sexo_pas = {'Hombre': 3, 'Mujer': 0}
coef_sexo_pad = {'Hombre': 2, 'Mujer': 0}

def get_coef_sexo_pas(sexo):
    return coef_sexo_pas.get(sexo, 0)

def get_coef_sexo_pad(sexo):
    return coef_sexo_pad.get(sexo, 0)

# Nuevos coeficientes para tus variables (puedes ajustar)
coef_imc_pas = 0.8
coef_imc_pad = 0.5

coef_diabetes_pas = 4  # Si Diabetes == 1
coef_diabetes_pad = 2

# Coeficientes para riesgo cardiovascular (num√©rico)
coef_riesgo_pas = 2
coef_riesgo_pad = 1

# Coeficiente para estr√©s num√©rico
coef_estres_pas = 1.5
coef_estres_pad = 1

# Coeficiente para consumo alcohol (num√©rico)
coef_alcohol_pas = 1
coef_alcohol_pad = 0.5

# Coeficiente para trigliceridos y colesterol total
coef_col_pas = 0.05
coef_tri_pas = 0.03

coef_col_pad = 0.03
coef_tri_pad = 0.02

pas_list = []
pad_list = []

for i, row in df.iterrows():
    base_pas = media_pas(row['Edad'])
    base_pad = media_pad(row['Edad'])
    
    pas = (base_pas
           + coef_imc_pas * (row['IMC'] - 25)
           + get_coef_sexo_pas(row['Sexo'])
           + coef_diabetes_pas * row['Diabetes']
           + coef_riesgo_pas * row['Riesgo_Cardiovascular_Num']
           + coef_estres_pas * row['Nivel_Estres_Num']
           + coef_alcohol_pas * row['Consumo_Alcohol_Num']
           + coef_col_pas * (row['Colesterol_Total'] - 180)
           + coef_tri_pas * (row['Trigliceridos'] - 150)
          )
    
    pad = (base_pad
           + coef_imc_pad * (row['IMC'] - 25)
           + get_coef_sexo_pad(row['Sexo'])
           + coef_diabetes_pad * row['Diabetes']
           + coef_riesgo_pad * row['Riesgo_Cardiovascular_Num']
           + coef_estres_pad * row['Nivel_Estres_Num']
           + coef_alcohol_pad * row['Consumo_Alcohol_Num']
           + coef_col_pad * (row['Colesterol_Total'] - 180)
           + coef_tri_pad * (row['Trigliceridos'] - 150)
          )
    
    pas += np.random.normal(0, 5)
    pad += np.random.normal(0, 3)
    
    pas = max(pas, 80)
    pad = max(pad, 50)
    
    pas_list.append(round(pas, 1))
    pad_list.append(round(pad, 1))

df['PAS'] = pas_list
df['PAD'] = pad_list
```

```python
# An√°lisis descriptivo de PAS y PAD
print("Resumen estad√≠stico de PAS:")
print(df['PAS'].describe())
print("\nResumen estad√≠stico de PAD:")
print(df['PAD'].describe())

# Visualizaci√≥n de las distribuciones
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.histplot(df['PAS'], bins=30, kde=True, color='royalblue')
plt.title('Distribuci√≥n de Presi√≥n Arterial Sist√≥lica (PAS)')
plt.xlabel('PAS (mmHg)')
plt.ylabel('Frecuencia')

plt.subplot(1,2,2)
sns.histplot(df['PAD'], bins=30, kde=True, color='orange')
plt.title('Distribuci√≥n de Presi√≥n Arterial Diast√≥lica (PAD)')
plt.xlabel('PAD (mmHg)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
